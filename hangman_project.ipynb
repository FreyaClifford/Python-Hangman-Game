```python
#Global counters to track statistics
number_of_games_played = 0
number_of_wins = 0
number_of_losses = 0
```

```python
#Allow the user to change the game difficulty
def game_difficulty():
    while True:   
        try:
            mode = input("choose a mode(easy/medium/hard) = ")
            if(mode.isalpha()):
                #Change the amount of guesses based on difficulty of the game
                if(mode == 'easy'):
                    number_of_guesses = 25

                elif(mode == 'medium'):
                    number_of_guesses = 20

                elif(mode == 'hard'):
                    number_of_guesses = 15

            print('number of guesses:', number_of_guesses)
            return number_of_guesses
        
        except:
            #Make an exception for misspelling to stop the game from crashing
            print("Please choose either easy, medium or hard")
        else:
            break
```

```python
#Asking the user to input a valid letter
def get_user_guess():
    while True:
        try:
            guessed = input("choose a letter")
            #Check if only one letter was guessed
            if(len(guessed)==1):
                return guessed 
            else:
                print("Please enter just one letter")
        #Use an except to prevent the game from crashing when recieving an invlaid input
        except ValueError:
            print("Please enter a letter")
```

```python
def play_round():
    
    global number_of_games_played, number_of_wins, number_of_losses
    
    number_of_games_played +=1
    number_of_guesses=game_difficulty()
    previous_guesses = []
    
    #Chooses a random word from the word bank and hides the letters
    secret_word = list(list_of_words[random.randint(0, len(list_of_words))])
    hidden_word = list(map(lambda x: '-',secret_word))
       
    while number_of_guesses > 0:
        print("You have %s guesses remaining" % number_of_guesses)
        print(hidden_word)
        guess = get_user_guess() 
        
        #Checks if letter has already been guessed
        if(guess in previous_guesses):
            print("You have already guessed this letter, please choose a different letter")
        else:
            #Add letter to previously guessed list and decrease the number of guesses remaining
            previous_guesses.append(guess)
            number_of_guesses = number_of_guesses - 1
            
            #Checks if the letter is in the word and returns with correct or incorrect   
            if(guess in secret_word):
                print("Correct")
                #Replace dashes with the correctly guessed letter
                for i in range(len(secret_word)):
                    if secret_word[i] == guess:
                        hidden_word[i] = guess
            else:
                print("Incorrect")
                
        #End game if full word has been guessed        
        if (hidden_word == secret_word):
            number_of_wins +=1
            print(hidden_word)
            print("You have guessed the full word correctly with %s guesses remaining! " % number_of_guesses)
            break
    
    #Reveal chosen word when out of guesses 
    if (hidden_word != secret_word):  
        number_of_losses +=1
        answer = ''.join(secret_word)
        print("You have run out of guesses, the word was: %s" % answer)   
```

```python
def main():
    
    print("Welcome to Hangman")
    
    while True:
        #Play a round of the game 
        play_round() 
        replay = input("Would you like to play again? (yes/no): ").strip().lower()
        if replay != "yes":
            #If the player chooses not to continue, end the game and show statistics
            print("Thank you for playing! Here are your statistics: ")
            print("Total games played: ",number_of_games_played)
            print("Number of wins: ", number_of_wins)
            print("Number of losses: ", number_of_losses)
            break
            
# This ensures the game only runs when this file is executed directly
if __name__ == "__main__":
    main()
```
